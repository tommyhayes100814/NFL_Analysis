#### Run a simulation for the run yards for a certain game, do this by the following: 
* Divide the plays into different categories based on down and position on the field
* Pull all games where that team has been in that "category" before
* Randomly select n values from that category (n is determined by the estimated amount of times they will be in that position during a certain game)
* Sum all the values together, then run the simulation 50-100 to get the final dist of values

```python
run_offense = (
    df.loc[df['play_type'] == 'run']
    [['game_id', 'posteam', 'down', 'ydstogo', 'yards_gained', 'goal_to_go', 'yardline_100']]
)

# filter out anywhere where down is null (two point conversions)
run_offense = run_offense.loc[~run_offense['down'].isna()]
```

```python
# break into certain categories that you can randomly assign
run_offense['run_category'] = np.where(
    run_offense['yardline_100'].between(90, 100), '90-100', np.where(
    run_offense['yardline_100'].between(70, 89), '70-89', np.where(
    run_offense['yardline_100'].between(50, 69), '50-69', np.where(
    run_offense['yardline_100'].between(30, 49), '30-49', np.where(
    run_offense['yardline_100'].between(11, 29), '11-29', np.where(
    run_offense['yardline_100'].between(0, 10), '0-10', 'ERROR'
    )))))
)

run_offense['category_id'] = run_offense['down'].astype(str) + '_' + run_offense['run_category']
```

```python
def simulate_game_offense(game_id, iters):
    game_df = run_offense.loc[run_offense['game_id'] == game_id]
    run_plays_category = (
        game_df.groupby(['category_id', 'posteam'])
               .agg({'game_id' : 'count'})
               .reset_index()
               .rename(columns={'game_id' : 'category_count'})
    )

    categories = run_plays_category['category_id'].unique()
    team_a_val, team_b_val = [], []
    original_teams = run_plays_category['posteam'].unique().tolist()
    for _ in range(iters): 
        final_df = pd.DataFrame()
        for category in categories: 
            category_df = run_plays_category.loc[run_plays_category['category_id'] == category]
            # category_df = run_plays_category.loc[run_plays_category['category_id'] == '1.0_50-69']
            teams = category_df['posteam'].unique().tolist()
            for team in teams: 
                # filter to the certain teams and categories
                team_df = category_df.loc[category_df['posteam'] == team]
                category_lookup = {
                    'posteam' : team_df['posteam'].tolist()[0], 
                    'category_count' : team_df['category_count'].tolist()[0]
                }

                for posteam, size in category_lookup.items(): 
                # now choose n random values and average them
                    choice_df = run_offense.loc[
                        ((run_offense['category_id'] == category) & 
                        (run_offense['posteam'] == category_lookup['posteam']))
                    ]

                    vals = np.random.choice(choice_df['yards_gained'].tolist(), 
                                            size=category_lookup['category_count'])
                    iteration = {
                        'game_id' : game_id, 
                        'category_id' : category, 
                        'offense_team' : category_lookup['posteam'], 
                        'mean_values' : np.mean(vals), 
                        'total_occurences' : category_lookup['category_count']
                    }

                i = pd.DataFrame(iteration, index=[0]).drop_duplicates()
                final_df = pd.concat([final_df, i], axis=0)

        final_df['total_yards'] = final_df['mean_values'] * final_df['total_occurences']
        teams = final_df['offense_team'].unique().tolist()
        team_a, team_b = teams[0], teams[1] 
        team_a = final_df.loc[final_df['offense_team'] == team_a]['total_yards'].sum()
        team_a_val.append(team_a)
        team_b = final_df.loc[final_df['offense_team'] == team_b]['total_yards'].sum()
        team_b_val.append(team_b)

    final_avg = {
         teams[0] : team_a_val, 
         teams[1] : team_b_val
    }
    
    return final_avg
```

```python
def plot_game(game_id, iters): 
    final = simulate_game_offense(game_id=game_id, iters=iters)
    team_a, team_b = list(final.keys())[0], list(final.keys())[1]
    
    actual = run_offense.loc[run_offense['game_id'] == game_id].groupby('posteam').agg({'yards_gained' : 'sum'}) 

    actual_vals = actual.to_dict()
    team_a_val = actual_vals['yards_gained'][team_a]
    team_b_val = actual_vals['yards_gained'][team_b]
    
    f, ax = plt.subplots(1, 2, figsize=(12, 6))
    sns.kdeplot(final[team_a], fill=True, ax=ax[0])
    ax[0].axvline(x=team_a_val, ymin=0, ymax=1, color='red', label='Actual')
    ax[0].set_title(f'Plot for {team_a}')
    sns.kdeplot(final[team_b], fill=True, ax=ax[1])
    ax[1].axvline(x=team_b_val, ymin=0, ymax=1, color='red', label='Actual')
    ax[1].set_title(f'Plot for {team_b}')
    plt.legend()
    plt.suptitle('Distrubution Plots for {0}'.format(game_id))
    plt.tight_layout()
    plt.show()
```

```python
# get 15 random games and see how well you do on average
random_games = list(np.random.choice(run_offense['game_id'].unique(), size=15))
for game_id in random_games: 
    plot_game(game_id=game_id, iters=75)
```
